```python
#Write a Python function that takes an alphanumeric string and returns the sum of all numeric parts.
 # (e.g., input: "abc12def3", output: 15)

def alphanumeric(s):
    total = 0
    current_number = ''

    for char in s:
        if char.isdigit():
            current_number += char  
        else:
            if current_number:
                total += int(current_number)  
                current_number = ''  

  
    if current_number:
        total += int(current_number)

    return total


input_str = "abc12def3"
print(alphanumeric(input_str))  
```

```python
#Write a Python function that takes an alphanumeric string and returns the sum of all numeric parts.
 # (e.g., input: "abc12def3", output: 15)

def alnumeric(str1):
    total=0
    char=""
    for x in str1:
        if x.isdigit():
            char+=x
        else:
            if char:
                total+=int(char)
                char=""
    if char:
        total+=int(char)
        return total

str1=input()
print(alnumeric(str1))
```

```python
def palindrome(n):
    temp=n
    rev=0
    while n>0:
        dig=n%10
        rev=rev*10+dig
        n=n//10
    if temp==rev:
        return True
    else:
        return False
n = int(input())
if palindrome(n):
    print("palindrome")
else:
    print("not a palindrome")
    
```

```python
class A:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def add(self):
        return self.x+self.y


    def sub(self):
        return self.y-self.x

class B(A):
    def __init__(self,x,y,z):
        super().__init__(x,y)
        self.z=z
    def mul(self):
        return self.x*self.y

s=A(10,4)
print(s.add())
s1=B(10,23,13)
print(s1.mul())
    


```
